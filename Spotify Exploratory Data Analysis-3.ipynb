```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data set
df_tracks = pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
#check for null values in the dataset
pd.isnull(df_tracks).sum()
```

```python
#check for data info
df_tracks.info()
```

```python
#find 10 least popular songs
sorted_df = df_tracks.sort_values("popularity", ascending = True).head(10)
sorted_df
```

```python
#descriptive statistics
df_tracks.describe().transpose()
```

```python
#10 most popular songs greater than 90
most_popular = df_tracks.query("popularity>90", inplace = False).sort_values('popularity', ascending = False).head(10)
most_popular
```

```python
#replace index with release date and replce in the original df
df_tracks.set_index("release_date", inplace =True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"],axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1,vmax = 1, center = 0,cmap = "inferno", linewidth = 1, linecolor = "Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#relationship between loudness and energy
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, x = "energy", y = "loudness", color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
#relationship between popularity and acousticness
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, x = "acousticness", y = "popularity", color = "y").set(title = "Popularity vs Acousticness Correlation")
```

```python
#create a new column called 'year' from release date
df_tracks["dates"]=df_tracks.index.get_level_values("release_date")
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of Songs per Year")
```

```python
total_duration = df_tracks.duration
figure_dimensions=(18,7)
figure,axis=plt.subplots(figsize=figure_dimensions)
figure=sns.barplot(x=years,y=total_duration,errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```
